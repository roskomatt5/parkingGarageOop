```python
class ParkingGarage:
    def __init__(self, tickets, parkingSpaces, currentTicket):
        self.tickets = tickets
        self.parkingSpaces = parkingSpaces
        self.currentTicket = currentTicket
        self.spot_taken = self.currentTicket['taken_spots'] = []
        self.ticket_taken = self.currentTicket['assigned_ticket'] = []
        self.is_paid = self.currentTicket['is_Paid'] = []

    def takeTicket(self):
        spaces = ", ".join(self.parkingSpaces)
        spot_selection = input(f'Hello! Here are the available spaces, please select one to continue: {spaces}. ')

        if  len(self.parkingSpaces) == 0:
            print('Sorry, there are no more spaces available')
        elif spot_selection not in self.parkingSpaces:
            print('That space is not available!')

        else:
            print(f'Here is your ticket "{self.tickets[0]}" Please Take note of it for payment processing later on.')
            self.is_paid.append(False)
            self.ticket_taken.append(self.tickets[0])
            self.spot_taken.append(spot_selection)
            self.parkingSpaces.remove(spot_selection)
            self.tickets.pop(0)
            self.spots_left = len(self.tickets)
            print(f'Happy Parking! There are {self.spots_left} spots left!')
    
    def payForParking(self):
        ticket_input = input('Please enter your ticket for payment here: ')
    
        if ticket_input not in self.ticket_taken:
            print('Space not assigned. Try Again.')
        else:
            ticket_index = self.ticket_taken.index(ticket_input)
            if ticket_input in self.currentTicket['assigned_ticket']:
                payment_input = float(input('Please input "15.00": '))
                if payment_input == 15.00:
                    self.is_paid[ticket_index] = True
                    print('Thank You for your payment! Please Exit the garage within 15 minutes to avoid further fees.')
                else:
                    print('Invalid payment amount. Try again.')
            else:
                print('Invalid entry. Try again.')

     
            
    def leaveGarage(self):
        ticket_input = input('Please insert ticket: ')
        ticket_index = self.ticket_taken.index(ticket_input)
        if self.currentTicket['is_Paid'][ticket_index] == True:
            print('Thank you, have a nice day!')
            self.ticket_taken.remove(ticket_input)
            self.parkingSpaces.append(self.spot_taken[ticket_index])
            self.spot_taken.pop(ticket_index)
            self.tickets.append(ticket_input)
            self.is_paid[ticket_index] = False
        elif self.currentTicket['is_Paid'][ticket_index] != True:
                payment_input = float(input('Please input "15.00": '))
                if payment_input == 15.00:
                    print('Thank you, have a nice day!')
                    self.is_paid[ticket_index] = True
                    self.ticket_taken.remove(ticket_input)
                    self.parkingSpaces.append(self.spot_taken[ticket_index])
                    self.spot_taken.pop(ticket_index)
                    self.tickets.append(ticket_input)
                else:
                    print('Invalid payment amount. Try again.')
        else:
            print('Invalid entry. Try again.')
            


run_garage = ParkingGarage(['1a','2b','3c'],['AA','BB','CC'],{})
#['AA','BB','CC']

def run():
    while True:
        response = input('What do you want to do? Take a ticket / pay / or exit? ')

        if response.lower() == 'take a ticket':
            run_garage.takeTicket()
            
        if response.lower() == 'pay':
            run_garage.payForParking()
            
        if response.lower() == 'exit':
            run_garage.leaveGarage()
    
run()
```
